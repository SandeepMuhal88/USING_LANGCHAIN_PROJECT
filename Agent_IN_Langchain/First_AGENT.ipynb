```python
from langchain_core.tools import Tool
```

```python
# Tools
tools = [
    Tool(
        name="calculator",
        func=lambda x: str(eval(x)),
        description="Useful for arithmetic."
    )
]
```

```python
# ReAct-style prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a ReAct agent. Use reasoning steps + actions."),
    ("human", "{input}")
])
```

```python
# Build ReAct agent (correct signature)
agent = create_react_agent(
    model=llm,
    tools=tools,
    prompt=prompt
)
```

```python
result = agent.invoke({"input": "What is 27 * 15?"})
print(result)
```
