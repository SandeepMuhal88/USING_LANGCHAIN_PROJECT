```python
from googleapiclient.discovery import build

API_KEY = "ENTER API KEY HERE"
VIDEO_ID = "Rni7Fz7208c"

youtube = build("youtube", "v3", developerKey=API_KEY)

# Step 1: List caption tracks
captions = youtube.captions().list(
    part="id,snippet",
    videoId=VIDEO_ID
).execute()

if not captions.get("items"):
    raise Exception("No captions available.")

# Prefer English captions
caption_id = None
for item in captions["items"]:
    if item["snippet"]["language"] in ["en-IN", "en"]:
        caption_id = item["id"]
        break

if caption_id is None:
    caption_id = captions["items"][0]["id"]

# Step 2: Download transcript
caption = youtube.captions().download(
    id=caption_id,
    tfmt="srt"
).execute()

print(caption["body"])
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from langchain_core.documents import Document
```

```python
video_id = "-j8PzkZ70Lg"
cookie_file_path = "www.youtube.com_cookies.txt" # Ensure this file exists

try:
    # Pass the cookie file here to bypass the IP block
    transcript_list = YouTubeTranscriptApi.get_transcript(
        video_id, 
        languages=['en'],
        cookies=cookie_file_path 
    )
    
    # Process into LangChain Document
    text_data = " ".join([t['text'] for t in transcript_list])
    
    doc = Document(
        page_content=text_data,
        metadata={"source": video_id, "url": f"https://www.youtube.com/watch?v={video_id}"}
    )
    
    print("Success! Transcript loaded.")
    print(doc.page_content[:200])

except Exception as e:
    print(f"Still failed: {e}")
```

```python
from langchain_community.document_loaders import YoutubeLoader

video_id = "Ilg3gGewQ5U"
video_url = f"https://www.youtube.com/watch?v={video_id}"

try:
    # Initialize the loader
    loader = YoutubeLoader.from_youtube_url(
        video_url, 
        add_video_info=True,  # Adds title, view count, etc. to metadata
        language=["en", "hi"], # Tries English first, then Hindi
        translation="en"      # Optional: Translate to English if not available
    )

    # Load the documents
    docs = loader.load()

    # Print the content
    print(docs[0].page_content)

except Exception as e:
    print(f"Error loading video: {e}")
```
