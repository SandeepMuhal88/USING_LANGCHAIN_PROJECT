```python
import browser_cookie3
from youtube_transcript_api import YouTubeTranscriptApi

video_id = "Rni7Fz7208c"

cookies = browser_cookie3.chrome()  # or .firefox()

transcripts = YouTubeTranscriptApi.list_transcripts(video_id, cookies=cookies)
t = transcripts.find_transcript(['en-IN'])  # manually created EN-IN
data = t.fetch(cookies=cookies)

full_text = " ".join([d['text'] for d in data])
print(full_text)
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

video_id = "Rni7Fz7208c"

# Step 1: List all transcripts
transcripts = YouTubeTranscriptApi.list_transcripts(video_id)

# Step 2: Pick manually created EN-IN transcript
t = transcripts.find_transcript(['en-IN'])

# Step 3: Fetch the transcript text
data = t.fetch()

# Step 4: Convert to single text
full_text = " ".join([entry['text'] for entry in data])

print(full_text)
```
